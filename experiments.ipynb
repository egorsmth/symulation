## world generation experiments

```python
import noise
import numpy as np
from skimage import io
import matplotlib.pyplot as plt


shape = (1024,1024)
scale = 100.0
octaves = 6
persistence = 0.5
lacunarity = 2.0

def getWorld(octaves = 6, persistence = 0.5, lacunarity = 2.0):
    world = np.zeros(shape)
    for i in range(shape[0]):
        for j in range(shape[1]):
            world[i][j] = noise.pnoise2(i/scale, 
                                        j/scale, 
                                        octaves=octaves, 
                                        persistence=persistence, 
                                        lacunarity=lacunarity, 
                                        repeatx=1024, 
                                        repeaty=1024, 
                                        base=0)
    return world

worlds = []
c = 0
final = 125
for o in range(1, 10, 2):
    for p in np.arange(0.1, 1.0, 0.2):
        print(c, "/", final)
        for l in np.arange(1.0, 10.0, 2.0):
            worlds.append(getWorld(o, p, l))
            c += 1

io.imshow_collection(worlds, cmap=plt.cm.gray)
```

```python
c = 0
for o in range(1, 10, 2):
    for p in np.arange(0.1, 1.0, 0.2):
        print(c, "/", final)
        for l in np.arange(1.0, 10.0, 2.0):
            name = "o-{}_p-{}_l-{}".format(o, p, l)
            io.imsave(name + ".png", worlds[c])
            io.imsave(name+"_color.png", add_color(worlds[c]))
            c += 1
```

```python
water = [65,105,225]
grass = [34,139,34]
wood = [0,100,0]
stone = [200, 200, 200]

def add_color(world):
    color_world = np.zeros(world.shape+(3,))
    for i in range(shape[0]):
        for j in range(shape[1]):
            if world[i][j] < -0.25:
                color_world[i][j] = water
            elif world[i][j] < 0.0:
                color_world[i][j] = grass
            elif world[i][j] < 0.25:
                color_world[i][j] = wood
            elif world[i][j] <= 1.0:
                color_world[i][j] = stone

    return color_world.astype(int)

color_world = add_color(world)
io.imshow(color_world)
```
